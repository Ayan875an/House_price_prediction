Train ml model

```python
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import SGDRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR

from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.neural_network import MLPRegressor
from sklearn.isotonic import IsotonicRegression

from xgboost import XGBRegressor
```

```python
#create object of every model
svr=SVR()
lr=LinearRegression()
sgdr=SGDRegressor()
knr=KNeighborsRegressor()
gpr=GaussianProcessRegressor()
dtr=DecisionTreeRegressor()
rfr=RandomForestRegressor()
gbr=GradientBoostingRegressor()
mlpr=MLPRegressor(max_iter=1000) # Increased max_iter to 1000
ir=IsotonicRegression()
xgbr=XGBRegressor()
```

```python
models ={"a":["Linear Regression", lr],
         "b":["SVR", svr],
         "c":["SGDRegressor", sgdr],
         "d":["KNeighborsRegressor", knr],
         "e": ["GaussianProcessRegressor", gpr],
         "f":["DecisionTreeRegressor", dtr],
         "g": ["GradientBoostingRegressor", gbr],
         "h":["RandomForestRegressor", rfr],
         "i": ["XGBRegressor", xgbr],
         "j": ["MLPRegressor", mlpr],
         #"k": ["IsotonicRegression", ir] #Removed IsotonicRegression because it only supports 1 feature
         }
```

```python
#Cross validation

from sklearn.model_selection import KFold, cross_val_score  # kfold= how many partion we want to create
from sklearn.metrics import make_scorer, r2_score # r2_score= from the metrics r2 model evaluation

def test_model(model, x_train=x_train, y_train=y_train):
    cv = KFold(n_splits = 7, shuffle=True, random_state = 45)
    r2 = make_scorer(r2_score)
    r2_val_score = cross_val_score(model, x_train, y_train, cv=cv, scoring = r2)
    score = [r2_val_score.mean()]
    return score
```

```python
models_score=[]
for model in models:
  print("Training model: ",models[model][0])
  score=test_model(models[model][1],x_train,y_train)
  print("model score: ",score)
  models_score.append([models[model][0],score[0]])
```

```python
models_score
```

```python
import numpy as np
import pandas as pd

# Find the best model based on the cross-validation score
best_model_name = max(models_score, key=lambda item: item[1])[0]
best_model = models[next(key for key, value in models.items() if value[0] == best_model_name)][1]


# Fit the best model
best_model.fit(x_train, y_train)

# Make predictions on test set
y_pred = best_model.predict(x_test)

# If target was log-transformed, reverse it
# y_pred = np.exp(y_pred)
"""Why log-transform the target?
        In many datasets (like House Prices), the target (SalePrice) is highly skewed (a few very large values).
        Training directly on skewed data can hurt performance.

Why reverse with np.exp()?
      When the model is trained on log(SalePrice), its predictions will also be in log scale.
      To get back the real values (SalePrice in dollars)

Example
Suppose:
Actual SalePrice = 200,000
Log-transformed = np.log(200000) ≈ 12.2
Your model predicts 12.1.
Without reversing → prediction looks like 12.1 (wrong scale).
With np.exp(12.1) ≈ 180,000 → now it’s a proper price again."""

# Round predictions for cleaner display
y_pred = np.round(y_pred, 2)

# Show first 10 predictions
print("Sample Predictions:")
display(y_pred[:10])

# (Optional) Compare with actual values if y_test is available
#if 'y_test' in locals():
#    comparison = pd.DataFrame({
#       "Actual": np.round(np.exp(y_test[:10]) if log_transformed else y_test[:10], 2),
#      "Predicted": y_pred[:10]
#   })
# print("\nSample Actual vs Predicted:")
# display(comparison)
```
